# Computing Points For Genius of Gaudí

![demo.png](demo.png)

We are going to compute some point positions. They are used in the beautiful proof of the equivalence of two definitions of the parabola, where one uses the conic section and the other the locus of points. The proof is attributed to Adolphe Quetelet and Germinal P. Dandelin.

This notebook computes the positions of some points found in Figure 2.2.4 of the book "Genius of Gaudí: Geometry and Architecture" by Claudi Alsina and Roger B. Nelsen.

```python
import sympy as sp
from sympy import Matrix
from sympy.geometry import Point3D, Plane
```

```python
# The origin of the coordinate system, the touching point of two cones
origin = Point3D(0, 0, 0)
```

We assume a right-hande coordinate with z-axis looking vertically upwards.

## Defining point $E$

```python

E_x, E_y, E_z = sp.symbols('E_x E_y E_z')

E = Point3D(E_x, E_y, E_z)
```

## Defining the sphere with center $c$

```python
c_z = sp.symbols('c_z')
sphere_center = Point3D(0, 0, c_z)
sphere_radius = sphere_center.distance(E)

eq = sp.Eq(E.distance(origin) / sphere_center.distance(origin), sp.cos(sp.rad(45)))

solutions_c_z = sp.solve([eq], [c_z], dict=True)
solutions_c_z
```

We assume $c_z < 0$:

```python
solution_c_z = solutions_c_z[0]
solution_c_z
```

## Defining point F

```python
F_x, F_y, F_z = sp.symbols("F_x F_y F_z")
F = Point3D(F_x, F_y, F_z)
```

## Defining plane $p$

Without losing generality, we can assume the plane normal of $p$ to have y-component 0. (Basically camera rotation.)

```python
p_normal_vector = Matrix([1, 0, -1])
p_normal_vector = p_normal_vector / p_normal_vector.norm()

p = Plane(F, normal_vector=p_normal_vector)

F_position_on_sphere = sp.Eq(Matrix(F.args), Matrix(sphere_center.args) + sphere_radius * Matrix(p.normal_vector) )
```

## Substituting actual values

While the points can move freely (with some constraints for convenience we set in this notebook), we can fix their positions by setting the z-coordinate of $E$ while giving some more constraints on other coordinate values.

```python
E_z_is_fixed = sp.Eq(E_z, -1) # This can be changed.

# For simplicity, we set its x-coordinate to 0 and E_y = E_z.
# We could constrain them in different ways, and this notebook should still work.
E_x_is_zero = sp.Eq(E_x, 0)
E_y_is_E_z  = sp.Eq(E_y, E_z)

solution_E = sp.solve([E_x_is_zero, E_y_is_E_z, E_z_is_fixed], [E_x, E_y, E_z], dict=True)

assert len(solution_E) == 1

solution_E = solution_E[0]
E.subs(solution_E)
```

```python
sphere_center.subs(solution_c_z).subs(solution_E)
```

```python
sphere_radius.subs(solution_c_z).subs(solution_E)
```

```python
# system = [eq.subs(solution_c_z).subs(solution_E) for eq in [F_is_on_p, F_is_on_sphere]]
system = [F_position_on_sphere.subs(solution_c_z).subs(solution_E)]

solutions_F = sp.solve(system, [F_x, F_y, F_z], dict=True)

assert len(solutions_F) == 1

solution_F = solutions_F[0]

F.subs(solution_F)
```
