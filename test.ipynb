```python
import sympy as sp
from sympy.geometry import Point3D, Plane
```

```python
origin = Point3D(0, 0, 0)
```

## Defining the point $E$

```python

E_x, E_y, E_z = sp.symbols('E_x E_y E_z')

E = Point3D(E_x, E_y, E_z)
```

## Defining the sphere

```python
c_z = sp.symbols('c_z')
sphere_center = Point3D(0, 0, c_z)

eq = sp.Eq(E.distance(origin) / sphere_center.distance(origin), sp.cos(sp.rad(45)))

solutions_c_z = sp.solve([eq], [c_z], dict=True)
solutions_c_z
```

We assume $c_z < 0$:

```python
solution_c_z = solutions_c_z[0]
solution_c_z
```

## Substituting actual values

```python
# For simplicity, we set its x-coordinate to 0. This is mathematically not necessary.
E_x_is_zero = sp.Eq(E_x, 0)
E_y_is_E_z  = sp.Eq(E_y, E_z)
E_z_is_minus_1 = sp.Eq(E_z, -1)

solution_E = sp.solve([E_x_is_zero, E_y_is_E_z, E_z_is_minus_1], [E_x, E_y, E_z], dict=True)

assert len(solution_E) == 1

solution_E = solution_E[0]
solution_E
```

```python
sphere_center = sphere_center.subs(solution_c_z).subs(solution_E)
sphere_center
```
